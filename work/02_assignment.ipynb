>Using the data below, answer the following questions:

>1) Which entities (top 5) had the largest population density in 2020? - 
>2) Which entities have more water area than land area?
>3) Which entities increased in population the most in the last 10 years?
>4) What state bird accounts for the largest population as of 2020? Land area?
>5) How many entities' largest city is their capital city?
>6) Which city has the largest percent drop from their largest city to their 5th largest? 100*(1st largest - 5th largest)/(1st largest)

```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
facts.head(5)
```

```python
facts.columns
```

```python
#1 - Which entities (top 5) had the largest population density in 2020? 
facts['Population Dencity 2020'] = facts['Pop_2020']/ facts['Area_land']
facts[['State','Area_land','Pop_2020','Population Dencity 2020']].sort_values(by='Population Dencity 2020',ascending=False).head(5)
```

```python
#2 - Which entities have more water area than land area?
facts['difference']= facts['Area_water'] >= facts['Area_land'] # difference is true if the area of water is grater then aear of land
facts[['State','Area_water','Area_land','difference']].sort_values(by='difference',ascending= False)
facts[facts['difference']==True][['State']]
```

```python
#3 - Which entities increased in population the most in the last 10 years?
facts['growth']=facts['Pop_2020']-facts['Pop_2010']
facts[['State','Pop_2020','Pop_2010','growth']].sort_values(by='growth',ascending=False)
```

```python
#4 - What state bird accounts for the largest population as of 2020? Land area?
facts[['State_bird','Pop_2010']].sort_values(by='Pop_2010', ascending=False).head(1)
```

```python
facts[['State_bird','Area_land']].sort_values(by='Area_land', ascending=False).head(1)
```

```python
#5 How many entities' largest city is their capital city?
facts['same'] = facts['Capital']==facts['City_1']
facts[['State','Capital','City_1','same']].sort_values(by='same', ascending = False)
SameCap = facts[facts['same']==True][["State"]].count()
print('The count is',SameCap)
```

```python
#6 Which city has the largest percent drop from their largest city to their 5th largest? 
#100*(1st largest - 5th largest)/(1st largest)
facts['percent difference']= ((facts['city_1_pop'] - facts['city_5_pop'])/facts['city_1_pop'])*100
facts[['State','percent difference']].sort_values(by='percent difference', ascending = False).head(1)
#percent different is the calculation of the difference between the largest city and the smallest in percentage form
```
